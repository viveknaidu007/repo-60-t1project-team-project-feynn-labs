```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../mcdonalds.csv")
```

```python
df.head()
```

```python
df.shape

#1453 rows and 15 columns
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
 #convert categorical columns to numeric
df.replace({"Yes": 1, "No": 0}, inplace=True)
```

```python
df.head()
```

```python
# convert 'VisitFrequency' to numerical values
visit_frequency_mapping = {"Once a month": 1, "Every three months": 3, "Once a week": 4}
df["VisitFrequency"] = df["VisitFrequency"].map(visit_frequency_mapping)
```

```python
# convert 'Gender' to numerical values
df["Gender"] = df["Gender"].map({"Female": 0, "Male": 1})
```

```python
df.head()
```

```python
# plot age distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# plot visit frequency distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='VisitFrequency')
plt.title('Visit Frequency Distribution')
plt.xlabel('Visit Frequency')
plt.ylabel('Count')
plt.show()
```

```python
# plot gender distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
